```python
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv(r'https://analityk.edu.pl/wp-content/uploads/2020/12/data.csv')
df['date'] = pd.to_datetime(df.date, format='%d/%m/%Y')
x=df['date']
y=df['Close']
```

```python
plt.plot(x,y)
```

```python
# zmiana koloru i stylu
fig, ax = plt.subplots(figsize=(10,8))
res = ax.plot(x, y)
line = res[0]  # dziki temu moemy korzysta z autouzupeniania
line.set_c('#d1371f')
line.set_linestyle('dashed')
```

```python
# Set the title
fig, ax = plt.subplots(figsize=(10,8))
res = ax.plot(x, y)
line = res[0]  # Once we have line we can use tab completion
line.set_c('#d1371f')
line.set_linestyle('dashed')
title = ax.set_title('Bitcoin Pricing')
```

```python
# Tweak title position
fig, ax = plt.subplots(figsize=(10,3))
res = ax.plot(x, y)
line = res[0]  # Once we have line we can use tab completion
line.set_c('#d1371f')
line.set_linestyle('dashed')
title = ax.set_title('Bitcoin Pricing')
title.set_position((.2,.7)) #.20% 70%
```

```python
# Customize tick locations
fig, ax = plt.subplots(figsize=(10,8))
res = ax.plot(x, y)
line = res[0]  # Once we have line we can use tab completion
line.set_c('#d1371f')
line.set_linestyle('dashed')
title = ax.set_title('Bitcoin Pricing')
title.set_position((.2,.7))

yax = ax.get_yaxis()
yax.set_ticks([-50, 500, 2000, 5000, 40000])
```

```python
# Customize tick locations
fig, ax = plt.subplots(figsize=(10,8))
res = ax.plot(x, y)
line = res[0]  # Once we have line we can use tab completion
line.set_c('#d1371f')
line.set_linestyle('dashed')
title = ax.set_title('Bitcoin Pricing')
title.set_position((.2,.7))

yax = ax.get_yaxis()
yax.set_ticks([-50, 500, 2000, 5000, 40000])
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
```

```python
fig, ax = plt.subplots(figsize=(10,8))
res = ax.plot(x, y)
plt.ylim(0,1000)
```

```python
fig, ax = plt.subplots(figsize=(10,8))
res = ax.plot(x, y)
ax.text(ax.get_xlim()[0], 2100, 'Bitcoin', clip_on=True)
```

```python
# Tweak text properties
fig, ax = plt.subplots(figsize=(10,8))
res = ax.plot(x, y)
# Need TTF/OTF fonts (TTC currently fail)
ax.text(ax.get_xlim()[0], 2100, 'Bitcoin', clip_on=True,
       family='Comic Sans MS', size=20)
# bbox is a mpl.patches.Rectangle
ax.text(ax.get_xlim()[0], 5700, 'Bitcoin', clip_on=True,
       family='Serif', size=20, style='italic', bbox={'facecolor': 'red', 'alpha': .5})
```
